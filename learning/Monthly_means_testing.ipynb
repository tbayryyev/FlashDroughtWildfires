```python
import netCDF4 as nc
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Setting up the variables and dataset
fn = 'Data.nc'
data = xr.open_dataset(fn)
SM = data['SoilMoist_S_tavg']
Temp = data['AvgSurfT_tavg']
ET = data['Evap_tavg']
time = data['time']
```

```python
# finding the 40th percentile of soil moisture at each raster
fourty_percentile = SM.quantile(0.4,dim='time')
fourty_percentile
```

```python
# checking whether the soil moisture value at each raster is less than the 40th percentile value we calculated above
SM_percentile_truth = SM<fourty_percentile
SM_percentile_truth
```

```python
# finding the monthly means for temperature
Temp_monthly_means = Temp.groupby("time.month").mean("time")
Temp_monthly_means
```

```python
# testing the temp monthly means at a specific point to see if the anomaly is actually being calculated based on the monthly mean
specific_point = Temp_monthly_means.sel(lat=36.125,lon=-121.125)
specific_point
```

```python
# testing the temp monthly means at a specific point to see if the anomaly is actually being calculated based on the monthly mean
anom = Temp.groupby("time.month") - Temp_monthly_means
anom.sel(lat=36.125,lon=-121.125)
```

```python
# testing the temp monthly means at a specific point to see if the anomaly is actually being calculated based on the monthly mean
Temp.sel(lat=36.125,lon=-121.125)
```
