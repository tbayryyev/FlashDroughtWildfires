```python
from netCDF4 import Dataset as NetCDFFile 
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap
import xarray as xr
```

```python
# the first netcdf file downloaded (1st day 2003-02-01)
day1 = xr.open_dataset('1.nc4')
SM1 = day1['SoilMoist_S_tavg']
```

```python
SM1.plot()
plt.title("2003-02-01 - Individual netcdf file")
```

```python
# the last netcdf file downloaded (last day 2021-12-31)
day6909 = xr.open_dataset('6909.nc4')
SM6909 = day6909['SoilMoist_S_tavg']
```

```python
SM6909.plot()
plt.title("2021-12-31 - Individual netcdf file")
```

```python
# Now to validate if correctly merged all the netcdf files into 1 large netcdf file
merged = xr.open_dataset('Data.nc')
```

```python
# Getting the first day of the larger netcdf file (first index)
SM_merged = merged['SoilMoist_S_tavg'][0]
```

```python
SM_merged.plot()
plt.title("2003-02-01 - merged netcdf file")
```

```python
# Getting the last day of the larger netcdf file (last index - 6908)
SM_merged_lastday = merged['SoilMoist_S_tavg'][6908]
```

```python
SM_merged_lastday.plot()
plt.title("2021-12-31 - Merged netcdf file")
```

The two maps from the individual files compared to the two maps from the merged file are identical.
Thus the file was merged correclty and validated here since the first and last days were validated.
